```python
#df = df.drop('Over18', axis=1)
#df = df.drop('Employee', axis=1)
df = df.drop('StandardHours', axis=1)
df = df.drop('EmployeeCount', axis=1)
```

```python
df.head()
```

```python
# Convert 'No' to 0 and 'Yes' to 1 in all columns
for column in df.columns:
    if df[column].dtype == object:  # Check if the column is of object type
        df[column] = df[column].replace({'No': 0, 'Yes': 1})  # Replace 'No' and 'Yes'
        # Convert all remaining strings to NaN which are ignored by df.corr()
        df[column] = pd.to_numeric(df[column], errors='coerce')

df.corr()
```

```python
plt.figure(figsize=(14,14))
sns.heatmap(df.corr(), annot=True, fmt='.0%')
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
for column in df.columns:
    if df[column].dtype == np.number:
        continue
    df[column] = LabelEncoder().fit_transform(df[column])
```
